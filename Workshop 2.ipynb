# Data analysis

The algorithms used in the analysis of all datasets are: 

* K-Nearest Neighbors (KNN)
* Naive Bayes (Gaussian)
* Decision tree (DTC)
* Multilayer Perceptron (MLP)

First we divide the datasets in Y (target column) and X (features columns). X and Y are list and each position of the list stores the data corresponding to one dataset.

```python
# All columns except target
X = [None] * 10
# Target column
Y = [None] * 10

# Fill X and Y for each dataset
for i, dataset in enumerate(datasets):
    X[i] = dataset.loc[:, dataset.columns != 'target'] 
    Y[i] = np.array(dataset.loc[:, dataset.columns == 'target']).ravel()
```

Here we analyze whether the classes of each dataset are balanced or not and determines the evaluation measure best suited to the datasets.

```python
import matplotlib.pyplot as plt

# Type of evaluation
scoring_choosed = [None] * 10

for i, dataset in enumerate(datasets):
    # To plot the proportion of classes, we first count the values of the 'target' column and transform them into a frame
    class_distribuition = pandas.DataFrame(dataset['target'].value_counts())
    
    # Plot classes
    plt.figure()

    # We use the frame created above to plot the frequency of each class in the dataset
    plot = seaborn.barplot(x=class_distribuition.index, y='target', data = class_distribuition)
    plot.set_title(datasets_names[i])

    # Verify distribution of target classes
    total_sum = sum(class_distribuition['target'])
    partition = total_sum/len(class_distribuition['target'])
    for value in class_distribuition['target']:
        if(0.9 * partition <= value <= 1.1 * partition):
            scoring_choosed[i] = "accuracy" 
        else:
            scoring_choosed[i] = "balanced_accuracy"
            break
```

## Classification

```python
# Import classification methods
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# Instantiate classifiers
DTC = DecisionTreeClassifier()
KNN = KNeighborsClassifier()
GNB = GaussianNB()
MLP = MLPClassifier()
```

Set all possible parameters for each classification method.

```python
DTC_parameters = {'criterion' : ['gini','entropy','log_loss'], 'random_state' : [42]}
KNN_parameters = {'n_neighbors' : [1, 3, 5, 7, 9, 11, 13], 'weights' : ['uniform', 'distance']}
GNB_parameters = {}
MLP_parameters = {'hidden_layer_sizes' : [(5), (8), (15), (5, 3), (8, 5), (10, 5)], 'max_iter' : [3000], 'random_state' : [42]}
```

Applying 10-fold cross validation (GridSearchCV) in the datasets.

```python
GS_DTC = [None] * 10
GS_KNN = [None] * 10
GS_GNB = [None] * 10
GS_MLP = [None] * 10

# Define grid search cv based on the paremeters defined above and corresponding evaluation method (scoring)
for i, choosed in enumerate(scoring_choosed):
    GS_DTC[i] = GridSearchCV(DTC, DTC_parameters, cv = 10, scoring = choosed)
    GS_KNN[i] = GridSearchCV(KNN, KNN_parameters, cv = 10, scoring = choosed)
    GS_GNB[i] = GridSearchCV(GNB, GNB_parameters, cv = 10, scoring = choosed)
    GS_MLP[i] = GridSearchCV(MLP, MLP_parameters, cv = 10, scoring = choosed)
```

```python
# Train and evaluate the classifier with the dataset
for i in range(len(datasets)):
    GS_DTC[i].fit(X[i], Y[i])
    GS_KNN[i].fit(X[i], Y[i])
    GS_GNB[i].fit(X[i], Y[i])
    GS_MLP[i].fit(X[i], Y[i])
```

Getting the results of training and evaluation.

```python
results_DTC = [None] * 10
results_KNN = [None] * 10
results_GNB = [None] * 10
results_MLP = [None] * 10

for i in range(len(datasets)):
    results_DTC[i] = pandas.DataFrame(GS_DTC[i].cv_results_)
    results_KNN[i] = pandas.DataFrame(GS_KNN[i].cv_results_)
    results_GNB[i] = pandas.DataFrame(GS_GNB[i].cv_results_)
    results_MLP[i] = pandas.DataFrame(GS_MLP[i].cv_results_)
```

Filter only the desired data from the results and sort by accuracy score.

```python
# Labels of data we want to analyse
view = ['params', 'mean_test_score','std_test_score', 'rank_test_score']

for i in range(len(datasets)):
    results_DTC[i] = results_DTC[i][view].sort_values(by='rank_test_score')
    results_KNN[i] = results_KNN[i][view].sort_values(by='rank_test_score')
    results_GNB[i] = results_GNB[i][view].sort_values(by='rank_test_score')
    results_MLP[i] = results_MLP[i][view].sort_values(by='rank_test_score')
```

Show the results.

```python
for i, dataset in enumerate(datasets_names):
    print(f'For dataset {dataset}:')
    print(f'\tThe accuracy of DTC classifier is: {results_DTC[i]["mean_test_score"][0] :.2%}')
    print(f'\tThe accuracy of KNN classifier is: {results_KNN[i]["mean_test_score"][0] :.2%}')
    print(f'\tThe accuracy of GNB classifier is: {results_GNB[i]["mean_test_score"][0] :.2%}')
    print(f'\tThe accuracy of MLP classifier is: {results_MLP[i]["mean_test_score"][0] :.2%}')
    print()
```

### Absolute difference

Analysis of results by selecting the two best algorithms of each dataset and, comparing the two, check the best one through the absolute difference.

First we show the raw data and sort the results by the mean test score.

```python
# Create list of lists
datasets_algs = [[] for _ in range(len(datasets))]

# Append results in the lists and show the mean test score and standard deviation test score of each method fro each dataset
for i, algs in enumerate(datasets_algs):
    algs.append({"name": "DTC", "results": results_DTC[i]})
    algs.append({"name": "KNN", "results": results_KNN[i]})
    algs.append({"name": "GNB", "results": results_GNB[i]})
    algs.append({"name": "MLP", "results": results_MLP[i]})
    algs.sort(key = lambda acc : acc["results"]["mean_test_score"][0], reverse = True)

    print(f'Dataset: {datasets_names[i]}')
    for alg in algs:
        print(f'\t{alg["name"]}: mean score: {alg["results"]["mean_test_score"][0]}, std: {alg["results"]["std_test_score"][0]}')
    print()
```

Select the two best results and calculate the mean error, standard deviation of errors and the absolute difference for each datasets.

```python
from math import sqrt

absolute_difference = [None] * 10

for i, algs in enumerate(datasets_algs):
    mean_err_A0 = 1 - algs[0]["results"]["mean_test_score"][0]
    std_A0 = algs[0]["results"]["std_test_score"][0]

    mean_err_A1 = 1 - algs[1]["results"]["mean_test_score"][0]
    std_A1 = algs[1]["results"]["std_test_score"][0]

    dif_mean_err = mean_err_A0 - mean_err_A1
    std = sqrt( (std_A0**2 + std_A1**2) / 2.0 )
    absolute_difference[i] = dif_mean_err / std
```

Show the results calculated above.

```python
for i, algs in enumerate(datasets_algs):
    print(f'Dataset: {datasets_names[i]}, absolute difference: {absolute_difference[i]}')
    if absolute_difference[i] > 0:
        if absolute_difference[i] >= 2:
            print(f'{algs[1]["name"]} performs better than {algs[0]["name"]} with 95% confidence')
        else:
            print(f'{algs[1]["name"]} performs better than {algs[0]["name"]}')
    else:
        if absolute_difference[i] <= -2:
            print(f'{algs[0]["name"]} performs better than {algs[1]["name"]} with 95% confidence')
        else:
            print(f'{algs[0]["name"]} performs better than {algs[1]["name"]}')
    print()
```
